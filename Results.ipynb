```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def graph(df, model):
    # Plot training & validation accuracy values
    plt.plot(df['acc'])
    plt.plot(df['val_acc'])
    plt.title(model + ' accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Test'], loc='upper left')
    plt.show()

    # Plot training & validation loss values
    plt.plot(df['loss'])
    plt.plot(df['val_loss'])
    plt.title(model + ' loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Test'], loc='upper left')
    plt.show()

    plt.plot(df['i_over_u'])
    plt.plot(df['val_i_over_u'])
    plt.title(model +' IoU')
    plt.ylabel('IoU')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Test'], loc='upper left')
    plt.show()
```

```python
df_1 = pd.read_csv('segnet_current_training.csv')
```

```python
df_1.head()
```

```python
graph(df_1, 'SegNet')
```

```python
df = pd.read_csv('atrous_current_training.csv')
```

```python
df.head()
```

```python
graph(df, 'AtrousResNet50')
```

```python
# Plot training & validation accuracy values
plt.plot(df['val_acc'])
plt.plot(df_1['val_acc'])
plt.title('SegNet vs Atrous ResNet50')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Atrous ResNet50', 'SegNet'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(df['val_loss'])
plt.plot(df_1['val_loss'])
plt.title('SegNet vs Atrous ResNet50')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Atrous ResNet50', 'SegNet'], loc='upper left')
plt.show()

plt.plot(df['val_i_over_u'])
plt.plot(df_1['val_i_over_u'])
plt.title('SegNet vs Atrous ResNet50')
plt.ylabel('IoU')
plt.xlabel('Epoch')
plt.legend(['Atrous ResNet50', 'SegNet'], loc='upper left')
plt.show()
```
